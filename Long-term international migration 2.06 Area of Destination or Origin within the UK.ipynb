```python
from databaker.framework import *
import pandas as pd 
```

```python
from pathlib import Path

sourceFolder = Path('in')
sourceFolder.mkdir(exist_ok=True)

%run lib/scrape_ons.ipynb

metadata = scrape('https://www.ons.gov.uk/peoplepopulationandcommunity/populationandmigration/' \
                  'internationalmigration/datasets/longterminternationalmigrationareaofdestinationororiginwithintheuktable206')
metadata
```

```python
inputFile = sourceFolder / 'data.xls'
response = session.get(metadata['fileURL'])
with open(inputFile, 'wb') as f:
  f.write(response.content)
tab = loadxlstabs(inputFile, sheetids='Table 2.06')[0]
```

```python
observations = tab.excel_ref("B15").expand(RIGHT).filter(contains_string("Estimate")).expand(DOWN).is_not_blank().is_not_whitespace()
observations
```

```python
CI = observations.shift(RIGHT)
CI
```

```python
AreaCode = tab.excel_ref('B14').expand(RIGHT).is_not_blank()
AreaCode
```

```python
Area = tab.excel_ref('B12').expand(RIGHT).is_not_blank()
Area
```

```python
Revision = tab.excel_ref('AJ13').expand(RIGHT).is_not_blank()
Revision
```

```python
Flow = tab.excel_ref('A').expand(DOWN).by_index([17,46,75])
Flow
```

```python
Year = tab.excel_ref('A19').expand(DOWN) - Flow - tab.excel_ref('A').expand(DOWN).by_index([45,74,103]) - tab.excel_ref('A104').expand(DOWN)  
Year
```

```python
Tag = tab.excel_ref('B15').expand(RIGHT).is_not_blank()
Tag
```

```python
Dimensions = [
            HDim(Year,'Year',DIRECTLY,LEFT),
            HDimConst('Geography', 'K02000001'),
            HDim(AreaCode,'Area of Destination or Origin',CLOSEST,LEFT),
            HDim(Area,'Area',CLOSEST,LEFT),
            HDim(Flow,'Flow',CLOSEST,ABOVE),
            HDimConst('Measure Type', 'Count'),
            HDimConst('Unit','People (thousands)'),
            HDim(CI,'CI',DIRECTLY,RIGHT),
            HDim(Revision,'Revision',DIRECTLY,ABOVE)
            ]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```
